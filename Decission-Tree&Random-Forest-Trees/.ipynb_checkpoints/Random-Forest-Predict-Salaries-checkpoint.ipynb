```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv').drop('Position', axis=1)
```

```python
X = dataset.iloc[:,0:1].values
y = dataset.iloc[:,1].values
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=350)
```

```python
regressor.fit(X,y)
```

```python
regressor.predict(6.5)
```

```python
# Visualizing the random forest

X_grid = np.arange(min(X), max(X), 0.01).reshape(-1,1)

plt.scatter(X,y, c='red')
plt.plot(X_grid,regressor.predict(X_grid), 'blue')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Truth or Bluff (Random Forest)')
plt.show()
```

```python
for i in np.arange(100,500,20):
    regressor = RandomForestRegressor(n_estimators=i)
    regressor.fit(X,y)
    pred = regressor.predict(6.5)
    plt.scatter(i,pred)
plt.xlabel('Number of Trees')
plt.ylabel('Predicted Salary')
plt.title('Number of trees vs. Predicted Salary')
plt.show()
```
